```python
import cv2
import sys
import scipy.signal as signal        
import scipy.fftpack as fftpack      
import skimage as ski               
from matplotlib import pyplot as plt
from skimage import img_as_float    
from skimage import img_as_ubyte     
import time
import pyPyrTools as ppt
from pyPyrTools import pyPyrUtils
import numpy as np
import pyPyrTools as ppt
import os
```

```python
#Reconstruct the image from the pyramid
def reconstruct(pyr,pind,levels = 5):
    
    reshaped_pyr = []
    for k in range(levels-1,-1,-1):
        startIndex = sum(np.prod(j) for j in pind[0:k])
        endIndex = ind - sum(np.prod(j) for j in pind[k+1:])
        reshaped_pyr.append((pyr[startIndex:endIndex]).reshape(pind[k]))                
    reshaped_pyr = reshaped_pyr[::-1]                                                  
    return reconPyr(reshaped_pyr)
```

```python
#Convert each channel to a linear pyramid. 
def linearize(pyramid,pind):
    nLevels = len(pyramid)
    se = sum([np.prod(i) for i in pind])
    pyr = np.zeros(se)
    pyrl = []


    for j in range(nLevels):
        shape = np.shape(pyramid[j])
        pyrl = pyrl +  (pyramid[j].reshape(shape[0]*shape[1])).tolist()  
       

    return np.asarray(pyrl)
```

```python
#Defining the Video Parameters
video = 'face'
yourpath = os.getcwd()
path = os.path.abspath(os.path.join(yourpath, os.pardir))+'/source/'+video
vid  = cv2.VideoCapture(path + '.mp4')
alpha = 150          
lambda_c = 10          
chromAttenuation = 0.1
fl = 0.5
fh = 1.            
samplingRate = 30.    
```

```python
#Operating on the first frame

[low_a,low_b] = signal.butter(1,fl/samplingRate,'low')   
[high_a,high_b] = signal.butter(1,fh/samplingRate,'low')


width, height = int(vid.get(cv2.CAP_PROP_FRAME_WIDTH)),int(vid.get(cv2.CAP_PROP_FRAME_HEIGHT))
nChannels = 3 
fps =  int(vid.get(cv2.CAP_PROP_FPS)) 
frame_count = int(vid.get(cv2.CAP_PROP_FRAME_COUNT)) 


startIndex  = 1                                                
end =  frame_count-10    
                                          
fourcc = cv2.VideoWriter_fourcc('M','J','P','G')               
writer = cv2.VideoWriter('butter'+str(video)+'fl_'+str(fl)+'fh_'+str(fh)+'.avi', fourcc, 30, (width, height), 1) 


vid.set(1,1) 
_, bgrframe = vid.read() 
rgbframe = bgrframe[:,:,::-1]
rgbframe = img_as_float(rgbframe)

#Converting to YIQ color space. 

YIQ = ski.color.rgb2yiq(rgbframe)
Y,I,Q  = YIQ[:,:,0], YIQ[:,:,1], YIQ[:,:,2]


#Creating the image pyramid for each of the channels.
 
pyY = ppt.Lpyr(Y).pyr
pyI = ppt.Lpyr(I).pyr
pyQ = ppt.Lpyr(Q).pyr



nLevels  = len(pyY)
pind = ([np.shape(pyY[i]) for i in range(nLevels)])  

#Linearizing the pyramid. 

pyY = linearize(pyY,pind)
pyI = linearize(pyI,pind)
pyQ = linearize(pyQ,pind)
   
pyr = np.asarray([pyY,pyI,pyQ])

lowpass1 = pyr
lowpass2 = pyr
pyr_prev = pyr

output = rgbframe

writer.write(rgbframe)  
```

```python

#Reconstruct funtion using pyPyrTools. It uses functions from the pyPyrTools library. 

def reconPyr(pyr):
       
                                               
        filt2 = 'binom5'                #The filter is binom5 filter I ahve used here. 
        edges = 'reflect1';             #The edges is reflect1. I have used this here. 
        maxLev = len(pyr)
        levs = range(0,maxLev)                 # The levels is range(0,maxLev)
        filt2 = pyPyrUtils.namedFilter(filt2)  #The named Filter filt2 . This has been finalized here. 
        res = []
        lastLev = -1
        for lev in range(maxLev-1, -1, -1):
            if lev in levs and len(res) == 0:
                res = pyr[lev]
            elif len(res) != 0:
                res_sz = res.shape
                new_sz = pyr[lev].shape
                filt2_sz = filt2.shape
                if res_sz[0] == 1:
                    hi2 = pyPyrUtils.upConv(image = res, filt = filt2,
                                            edges = edges, step = (2,1), 
                                            stop = (new_sz[1], new_sz[0])).T
                elif res_sz[1] == 1:
                    hi2 = pyPyrUtils.upConv(image = res, filt = filt2.T,
                                            edges = edges, step = (1,2), 
                                            stop = (new_sz[1], new_sz[0])).T
                else:
                    hi = pyPyrUtils.upConv(image = res, filt = filt2, 
                                           edges = edges, step = (2,1), 
                                           stop = (new_sz[0], res_sz[1]))
                    hi2 = pyPyrUtils.upConv(image = hi, filt = filt2.T, 
                                            edges = edges, step = (1,2),
                                            stop = (new_sz[0], new_sz[1]))
                if lev in levs:
                    bandIm =  pyr[lev]
                    bandIm_sz = bandIm.shape
                    res = hi2 + bandIm
                else:
                    res = hi2
        return res                           
                
   
```

```python
startIndex = 1
end = 200
```

```python
#Operating on frames from 2 to 100. 
for i in range(startIndex+1,end):                           
    vid.set(1,i)
    print "frame", i,"of",end
    _, bgrframe = vid.read() 
    rgbframe = bgrframe[:,:,::-1]
    rgbframe = img_as_float(rgbframe)
    YIQ = ski.color.rgb2yiq(rgbframe)
    Y,I,Q  = YIQ[:,:,0], YIQ[:,:,1], YIQ[:,:,2]
    
    pyY = ppt.Lpyr(Y).pyr
    pyI = ppt.Lpyr(I).pyr
    pyQ = ppt.Lpyr(Q).pyr
    
    nLevels  = len(pyY)
    
    pind = ([np.shape(pyY[i]) for i in range(nLevels)])
    
    pyY = linearize(pyY,pind)
    pyI = linearize(pyI,pind)
    pyQ = linearize(pyQ,pind)
    
    pyr = np.asarray([pyY,pyI,pyQ])

    lowpass1 = (-high_b[1]*lowpass1 + high_a[0]*pyr + high_a[1]*pyr_prev)/high_b[0]
    lowpass2 = (-low_b[1]*lowpass2 + low_a[0]*pyr + low_a[1]*pyr_prev)/low_b[0]
    filtered =  lowpass1-lowpass2 
    
    pyr_prev = pyr                                               
    ind = len(pyr[0])                                           
   
    delta = lambda_c/8./(1+alpha)                                                           
    exaggeration_factor = 2                                       
    lambd = (width^2+height^2)/3.

    for l in range(nLevels-1,-1,-1):
        startIndex = sum(np.prod(j) for j in pind[0:l])
        endIndex = ind - sum(np.prod(j) for j in pind[l+1:])
        currAlpha = lambd/delta/8. - 1
        currAlpha = currAlpha*exaggeration_factor;
        indices = range(startIndex,endIndex)                   
        
        if(l == nLevels - 1 or l==0):
            filtered[:,indices] = 0.                                                                       
        elif (currAlpha>alpha):
            filtered[:,indices] = alpha*filtered[:,indices]           
        
        else:
            filtered[:,indices] = currAlpha*filtered[:,indices]           

        lambd = lambd/2.
   

    #reconstructing the individual levels into a pyramid. 
    #chromAttenuation assigned to the I,Q of the pyramid. 

    output[:,:,0] = reconstruct(filtered[0,:],pind,nLevels)                      
    output[:,:,1] = reconstruct(filtered[1,:],pind,nLevels)
    output[:,:,2] = reconstruct(filtered[2,:],pind,nLevels)
    output[:,:,1] = chromAttenuation*output[:,:,1]
    output[:,:,2] = chromAttenuation*output[:,:,2]
    
    rgb_out = ski.color.yiq2rgb(output)
    
    rgbframe = rgbframe + rgb_out
    output_final = rgbframe[:,:,::-1]
    
    output_final[output_final>1] = 1
    output_final[output_final<-1] = -1
    
    output_final  = img_as_ubyte(output_final)
    writer.write(output_final)
    
writer.release()

  
```
